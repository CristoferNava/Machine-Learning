```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = pd.Series([1, 2, 2, 3, 4, 4, 5, 6])
Y = pd.Series([2, 3, 4, 4, 4, 6, 5, 7])
```

```python
plt.scatter(X, Y, c='#ef5423', label='Puntos')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
